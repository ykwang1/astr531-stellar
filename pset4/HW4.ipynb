# Q 20.2

### a.
From equation 20.10:
$$ \frac{L}{L_\odot} \approx 5.2 \times 10^{10} \frac{M}{M_\odot}\mu_{ion}^{-7/5}\Big(\frac{t}{yr}\Big)^{-7/5}$$
Thus at any time t, $\frac{L_{\mu_1}}{L_{\mu_2}} \approx \Big(\frac{\mu_1}{\mu_2}\Big)^{-7/5}$. As a result, the He rich white dwarf will be ~0.143 times as faint as a H white dwarf and a C-rich WD will be ~0.03 times as faint as a H-rich WD of the same age.

Note that this assumes that the masses of the WDs are identical, which would likely not be the case in real life, since more massive WDs would likely be composed of heavier elements.

### b.
The reason WDs composed to heavier elements cool more quickly (and are therefore less luminous, since almost all WD radiation is thermal radiation of energy of these ions) has to do with the specific heat capacity of the ions, $C_V$, which affects the rate at which energy is lost $\frac{dT}{dt}$. Because $C_V \propto 1/m_{ion}$, the heavier ions have a lower specific heat capacity and thus the rate at which these WDs lose energy is greater than H-rich WDs.

# Q 23.2
From 8.21:
$$ T_c \simeq \frac{\mu m_H}{k}\frac{GM}{R}$$
The original expression for $T_c \propto \rho_c^{1/3}$ was derived from the above expression. However, from the full form of the equation (8.22) we see that this depends on a constant core mass:
$$ T_c \approx \frac{m_H \mu}{k}GM_c^{2/3}\rho^{1/3}$$
Where this is just 8.21 with the density factored out. Since **each successive phase of fusion occurs in a smaller and smaller region of the core**, the effective mass in this equation is also changing as well, which is not taken into account by the simple proportionality. As a result, the temperature of the (smaller) core is less than predicted by the $T_c\propto\rho^{1/3}$ relation, by a factor of $(M_{c,1}/M_{c,2})^{2/3}$ where $M_{c,i}$ is the portion of the where fusion of element $i$ is taking place.

# Q 25.1
### a.
The mass loss rate of massive stars is dominated by radiation pressure, which depends on the surface temperature of the star. Because a rapidly rotating star has oblate surfaces of constant effective gravity, the star iself has a greater temperature gradient near the poles than the equator. This results in a hotter surface and greater radiative flux near the poles than the equator of rapidly rotating massive stars; this is the Von Zeipel effect. As a result, the mass loss rate is also greater nearer the poles than the surface of the star.

However, if the effective temperature at the equator is low enough (under 21000K) and the escape velocity and surface effective gravity is also low enough, mass loss at the equator may greatly exceed mass loss elsewhere. This is the bi-stability jump and occurs because the ions that drive the line driven wind are in a different (i.e. in a different ionization) under ~21000K.

### b.
Beause of the oblate shape of the surfaces of equal surface gravity, the surface gravity near the equator is less than the surface gravity near the poles. Because this surface gravity determines the escape velocity, and because we approximate the terminal wind velocity to be the escape velocity, the terminal wind velocity is highest at the poles and smallest at the equator.

### c.
The density of the wind at any given point is proportional to mass loss divided by velocity: $\rho \propto \dot{M}/v$. We established that the terminal velocity of the wind under 21000K is about half that over 21000K (see 15.2.2), due to the differences in ionizations--this is the bistability jump. Since the temperature at the equator is cool enough that the optical depth of the wind becomes greater and more opaque in the UV, which means that the ions driving the wind by line radiation pressure are in lower ionization states (Lamers, 1991). This also increases the mass loss rate as more momentum is transfered to these ions, increasing the mass loss.

As a result, the decreased wind velocity and increased mass loss rate both contribute to a greater density at the equator.
